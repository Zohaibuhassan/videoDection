```python
from ultralytics import YOLO
from google.colab import files
import os
import subprocess
from IPython.display import HTML
from base64 import b64encode
```

```python
print("üëâ Upload your video file (e.g., .mp4 or .avi)...")
uploaded = files.upload()
video_path = list(uploaded.keys())[0]
print(f"üìÇ Uploaded: {video_path}")
```

```python
model = YOLO("yolov8n.pt")  # Or use yolov8s.pt / yolov8m.pt for better accuracy
results = model.predict(source=video_path, save=True, conf=0.4)
```

```python
# Path to output video
output_dir = "runs/detect/predict"
output_files = os.listdir(output_dir)
output_video_path = os.path.join(output_dir, output_files[0])
print("‚úÖ Output video:", output_video_path)
```

```python
# Convert to mp4 if it is .avi or other non-browser format
if output_video_path.endswith(".avi"):
    output_mp4 = output_video_path.replace(".avi", ".mp4")
    print("üîÑ Converting AVI to MP4...")
    subprocess.call([
        'ffmpeg', '-i', output_video_path, '-vcodec', 'libx264', '-f', 'mp4', output_mp4
    ])
else:
    output_mp4 = output_video_path
```

```python
# Display the output video inline in notebook
mp4 = open(output_mp4, 'rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()

HTML(f"""
<video width=640 controls>
    <source src="{data_url}" type="video/mp4">
</video>
""")
```
